```
import math
def distance_from_origin(a, b, c):
  return abs(c)/math.sqrt(a**2+b**2)
```

```
def distance_btw_parellel_lines(a1, b1, c1, a2, b2, c2):
  c=abs(c1-c2)
  a=a1
  b=b1
  return abs(c)/math.sqrt(a**2+b**2)
```

```
#step function

X=np.linspace(start=-10, stop=10, num=100)
Y=np.where(X>0, 1, -1)
plt.plot(X, Y)

```

```
# Plotting a 3D surface
from mpl_toolkits.mplot3d import Axes3D # Importing this is necessary when you plot in 3D
a = 1
b = 1
c = 0
d = 10
##############################
xx = np.linspace(-10, 10, 20)
zz = np.linspace(-10, 10, 20)
X, Z = np.meshgrid(xx, zz)
yy = (-a*X - c*Z - d) * 1.0 / b
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, Z, yy)
```

```
#2Dguassian
xx=np.linspace(start=-10, stop=10, num=100)
yy=np.linspace(start=-10, stop=10, num=100)
X, Y = np.meshgrid(xx, yy)
zz=np.exp(-X**2-Y**2)
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, Y, zz)
```

Write code to plot square, rectangle and ellipse

```

#ELLIPSE

a=4
b=5
X=np.linspace(-a, a, 200)
Y1=np.sqrt(b**2*(1-X**2/a**2))
Y2=-np.sqrt(b**2*(1-X**2/a**2))

plt.plot(X, Y1)
plt.plot(X, Y2)
```

```
#SQUARE

a=4
b=4
X=np.linspace(-a, a, 200)
Y=np.linspace(-b, b, 200)
Y1=a+0*X
Y2=-a+0*X
X1=b+0*Y
X2=-b+0*Y

plt.plot(X, Y1)
plt.plot(X, Y2)
plt.plot(X1, Y)
plt.plot(X2, Y)
```

```
#Rectangle

a=8
b=4
X=np.linspace(-b, b, 200)
Y=np.linspace(-a, a, 200)
Y1=a+0*X
Y2=-a+0*X
X1=b+0*Y
X2=-b+0*Y

plt.plot(X, Y1)
plt.plot(X, Y2)
plt.plot(X1, Y)
plt.plot(X2, Y)
```
